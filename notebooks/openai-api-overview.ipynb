# OpenAI API Overview

he Responses API is the new core primitive: it replaces the old chat completions endpoint and is “OpenAI’s most advanced interface for generating model responses,” supporting tools, file search, code interpreter, and conversations as first-class concepts.

 In class, you can frame it as “the engine” that everything else runs on: whether you’re building a simple script or a complex agent, you’re ultimately orchestrating calls to the Responses API (sometimes directly, sometimes via higher-level tooling like the Agents SDK).

[Inference] OpenAI’s AgentKit / Agents SDK then becomes “the agent layer on top of Responses.” The Agents docs describe a modular toolkit that lets a model use tools, maintain state, and coordinate multi-step workflows, and they explicitly call out Agent Builder as a visual canvas for composing these workflows as nodes and connections.


Responses is how the model thinks and acts step-by-step, AgentKit is how you design and host those behaviors in code, and Agent Builder is how non-experts (or you, when teaching) can draw the same flows instead of hand-coding them. A nice storyline for students is: “We start with raw Responses calls, then move up to Agents SDK to structure multi-step behavior, and finally show how the same ideas are expressed visually in Agent Builder.”

[Inference] Finally, ChatKit is “the front-end skin” for all of this—the docs describe it as the best way to build agentic chat experiences, with pre-built widgets and components to embed your agents in a product UI.
